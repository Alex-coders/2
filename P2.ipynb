```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()
graph.add_edge(0, 1, weight=1)
graph.add_edge(0, 2, weight=3)
graph.add_edge(1, 3, weight=5)
graph.add_edge(2, 3, weight=1)
graph.add_edge(2, 4, weight=2)
graph.add_edge(3, 4, weight=1)
graph.add_edge(3, 5, weight=2)
graph.add_edge(4, 5, weight=4)

start_node = 0
goal_node = 5

shortest_path = nx.shortest_path(graph, source=start_node, target=goal_node, weight='weight')
shortest_path_length = nx.shortest_path_length(graph, source=start_node, target=goal_node, weight='weight')

if shortest_path:
    print(f"Minimum cost from {start_node} to {goal_node} is = {shortest_path_length}")
    edge_colors = ['red' if (u, v) in zip(shortest_path, shortest_path[1:]) else 'gray' for u, v in graph.edges()]
    pos = nx.spring_layout(graph)
    nx.draw(graph, pos, with_labels=True, node_color='lightgray', edge_color=edge_colors, arrows=True)
    nx.draw_networkx_edges(graph, pos, edgelist=[], width=2.0, alpha=0.5, edge_color='blue')
    plt.show()
else:
    print(f"No path exists from {start_node} to {goal_node}.")
```

```python
import networkx as nx
import matplotlib.pyplot as plt


def dist(a, b):
    (x1, y1) = a
    (x2, y2) = b
    return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5


G = nx.grid_graph(dim=[3, 3])  
nx.set_edge_attributes(G, {e: e[1][0] * 2 for e in G.edges()}, "cost")
path = nx.astar_path(G, (0, 0), (2, 2), heuristic=dist, weight="cost")
length = nx.astar_path_length(G, (0, 0), (2, 2), heuristic=dist, weight="cost")
print("Path: ", path)
print("Path length: ", length)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="#f86e00")
edge_labels = nx.get_edge_attributes(G, "cost")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
